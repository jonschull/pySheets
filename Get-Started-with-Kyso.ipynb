Personally, bokeh is my favourite!

But here are some basic plots of simple python arrays in each of the libraries above that you can start using to beautiful visualizations on Kyso:

````python
import numpy as np
x = np.linspace(0, 1, 200)
y = np.random.randn(200)+10
````

### Plotly

````python
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go

plotly.offline.init_notebook_mode(connected=True)

trace = go.Scatter(
    x = x,
    y = y,
    mode = 'lines+markers',
    name = 'lines+markers'
)

data = [trace]

layout = go.Layout(title="Plotly")

fig = go.Figure(data=data,layout=layout)

iplot(fig)
````

### Bokeh

````python
from bokeh.plotting import figure, output_notebook, show
output_notebook()

p = figure(plot_width=800, plot_height=400, title="Bokeh")

p.line(x, y, line_width=2)
p.circle(x, y, fill_color="white", size=8)

show(p) 
````

### Matplotlib

````python
%config InlineBackend.figure_format = 'retina'
import matplotlib as mpl
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Matplotlib')
````

### Matplotlib using Seaborn

````python
import seaborn as sns; sns.set()
ax = sns.lineplot(x=x, y=y)
ax.set_title("Seaborn")
````

## Markdown

Use hashtags followed by a blank space for notebook titles and section headings.

<mark># Title (H1)</mark>

<mark>## Major Headings (H2)</mark>
    
<mark>### Subheadings (H3)</mark>

<mark>#### 4th Level Subheadings (H4)</mark>

And so on..

You can wrap markdown in _ _ or * * to make italic or __ __ &  ** ** for bold.

### Lists

Use the dash sign ( - ) with two spaces after it or a space, a dash, and a space ( - ), to create a circular bullet. To create a sub bullet, use a tab followed by a dash and two spaces. You can also use an asterisk or a plus sign instead of a dash.

* Unordered lists can use Asterisks.
- Dashes.
+ Or the Plus Sign.

For a numbered list start with 1. followed by a space and the numbering will start auotmatically. Start each line with some number and a period, then a space. Tab to 
indent and get subnumbering.

1. First ordered list item.
2. Another item
  * Unordered sub-list. 
1. Actual numbers don't matter, just that it's a number.
  1. Ordered sub-list.
4. And another item.
